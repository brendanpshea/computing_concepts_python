# What is Artificial Intelligence?
### Brendan Shea, PhD

**Artificial Intelligence (AI)** refers to the simulation of human intelligence in machines designed to think and act like humans. This field of computer science aims to create systems capable of tasks that would typically require human intellect. These tasks include *learning* (acquiring information and the rules for using the information), *reasoning* (using the rules to reach approximate or definite conclusions), and *self-correction*.

A prime example of AI in action is the development of chess-playing programs. Early versions operated on brute-force calculations of possible moves, but modern AI chess programs, like *Deep Blue*, use advanced strategies and learning capabilities, illustrating the evolution of AI. Another example is **natural language processing** (NLP), seen in voice recognition systems and chatbots. NLP involves the computer's ability to understand, interpret, and respond to human language in a meaningful way.

AI is often categorized into two types: **narrow** or **weak AI**, which is designed for a specific task (like facial recognition or internet searches), and **general** or **strong AI**, which has the broader ability to apply human-level (or greater) intelligence to any problem, much like a human brain. Currently, all existing AI systems fall into the narrow AI category, meaning they perform well in certain restricted domains, but don't do as well outside those domains.

The development of AI involves various subfields, including **machine learning**, where algorithms are trained to make classifications or predictions, and **deep learning**, which involves teaching computers to learn by processing inputs through layers in a way analogous to the human brain. The advancements in these areas continue to push the boundaries of AI, making it a rapidly evolving and exciting field.

## A Little History of AI From Descartes to the Present
The history of AI and ideas about artificial intelligence span centuries, intertwining philosophical inquiries with technological advancements.

- **17th Century - Ren Descartes**. The journey begins with philosopher Ren Descartes, who, in the 17th century, pondered the possibility of machines thinking and acting like humans. His famous assertion "Cogito, ergo sum" (I think, therefore I am) laid the groundwork for later debates about consciousness and artificial intelligence.

- **19th Century - Ada Lovelace**. Fast forward to the 19th century, where Ada Lovelace, often considered the first computer programmer, speculated on **Charles Babbage**'s Analytical Engine. Lovelace famously suggested that the machine had the potential to manipulate symbols in accordance with rules and that it could do more than mere number crunching, hinting at early concepts of AI.

- **Mid-20th Century - Turing Test**. Alan Turing, in the mid-20th century, introduced the Turing Test as a criterion of intelligence, a fundamental concept in the philosophy of AI. This test assesses a machine's ability to exhibit intelligent behavior indistinguishable from that of a human.

- **1950s - Birth of AI**. The term "Artificial Intelligence" was first coined by **John McCarthy** in 1955. The 1956 Dartmouth Conference is often considered the birth of AI as a field of study. Here, researchers began exploring problem-solving and symbolic methods.

- **1960s - Expert Systems**. The 1960s witnessed the rise of expert systems, computer programs that mimic the decision-making abilities of a human expert. This period also saw advancements in machine learning and neural networks.

- **1980s - AI Winter**. The limitations of early AI systems led to the first "AI Winter" in the 1980s, a period of reduced funding and interest in AI research.This has happened pretty regularly in the history, as nearly every generation (from Descartes and Lovelace up to now) has tended to underestimate just how "tough" the problem is.

- **Late 20th Century - Internet and Increased Computing Power**. The proliferation of the internet and a significant increase in computing power revived interest in AI. **Machine learning**, especially, benefited from the ability to access and process large datasets.

- **21st Century - Deep Learning and Big Data**. The early 21st century saw breakthroughs in deep learning, substantially improving the performance of AI in tasks like image and speech recognition. The availability of big data and increased computational power have led to AI systems surpassing human performance in specific domains.

- **Present Day**. Today, AI is an integral part of daily life, influencing areas such as healthcare, finance, transportation, and personal assistants. Ethical and governance issues, such as AI bias, transparency, and the impact on employment, are at the forefront of current discussions in the field.


### How Does Machine Learning Differ from Traditional Programming?

The distinction between **Machine Learning (ML)** and Traditional Programming lies fundamentally in their approach to problem-solving and task execution. In traditional programming, a programmer writes explicit *instructions* or *code* to perform a specific task, based on logical sequences and defined algorithms. The computer then executes these instructions. A classic example is a sorting algorithm, where the programmer defines the exact steps to sort a list of items.

Conversely, machine learning, a subset of AI, involves teaching a computer system to make decisions or predictions based on data. Here, the focus is on developing **algorithms** that can *learn* from and make decisions or predictions based on data. In ML, the programmer does not explicitly code the decision-making process. Instead, they create algorithms that can learn from and adapt to data. For instance, in an email filtering system, instead of explicitly coding rules to identify spam, an ML model is trained on a dataset of emails labeled as 'spam' or 'not spam', and it learns to classify incoming emails.

One key component of machine learning is the **training process**. ML algorithms are trained using large sets of data, where they 'learn' patterns and features from this data. The 'learning' could be **supervised** (the algorithm is trained on a labeled dataset), **unsupervised** (the algorithm learns from an unlabeled dataset, identifying patterns and structures), or **reinforcement learning** (the algorithm learns by receiving feedback in the form of rewards or penalties).

A real-world example of machine learning is **recommendation systems** used by streaming services like Netflix or Spotify. These systems analyze your past behavior and the behavior of similar users to recommend movies or songs you might like. The system isn't programmed with explicit rules to make these recommendations but learns from the accumulated data.

### What Are Neural Networks and How Do They Work?

**Neural Networks** are a fundamental concept in AI and machine learning, inspired by the structure and function of the human brain. They are algorithms designed to recognize patterns and interpret sensory data through machine perception, labeling, and clustering raw input.

The basic building block of a neural network is the **neuron**, an analogy to biological neurons in the human brain. A neuron in a neural network receives input, processes it, and produces an output. In the context of neural networks, these neurons are arranged in layers: **input layers** (receiving data), **hidden layers** (processing the data), and **output layers** (producing the final result).

Data flows through the network from the input layer to the output layer. Each neuron in one layer is connected to neurons in the next layer. These connections have **weights**, which are adjusted during the training process. The strength of these weights determines how much influence one neuron has on another.

Neural networks learn by adjusting these weights based on the feedback they receive in the learning process. This is typically done using a method called **backpropagation**, where the network adjusts its weights to minimize the difference between its actual output and the desired output.

A classic example of a neural network application is **image recognition**. In this scenario, the input layer receives the raw pixel data from images. The hidden layers process this data, identifying patterns and features (like edges, shapes, and textures). Finally, the output layer classifies the image (for example, as a cat, dog, or something else).

Neural networks are especially potent in tasks involving complex pattern recognition, and they form the backbone of **deep learning**, where networks with many hidden layers can process vast amounts of data with a high degree of accuracy.

## Case Study: Decision Trees and the Wisconsin Breast Cancer Dataset


In this case study, we will be delving into the domain of medical informatics - a field at the intersection of information science, computer science, and healthcare. Specifically, we'll explore the realm of breast cancer diagnosis, a critical area of study given that breast cancer is one of the most common cancers among women worldwide.

Our focus will be on a machine learning algorithm known as the Decision Tree and we'll use it to classify whether a breast mass is benign (non-cancerous) or malignant (cancerous). This use-case clearly illustrates the potential of machine learning in facilitating healthcare professionals' decision-making processes and ultimately improving patient outcomes.

The dataset we'll be using is the Wisconsin Breast Cancer dataset, a commonly-used dataset in machine learning for binary classification problems. This dataset consists of measurements from digitized images of fine needle aspirates (FNA) of breast masses. Features are computed from these images, resulting in numerical values that describe characteristics of the cell nuclei present in the images. The 'target' in our dataset indicates whether the mass is malignant (cancerous) or benign (non-cancerous).

### Decision Trees
**Decision Trees** are a type of supervised learning algorithm that is mostly used for classification problems. Visually, it represents decisions and decision-making processes that mirror human thinking ability. In a Decision Tree, each internal node represents a "test" on an attribute, each branch represents the outcome of the test, and each leaf node represents a class label (a decision taken after testing all attributes). The paths from root to leaf represent classification rules.

The beauty of Decision Trees lies in their simplicity and interpretability. Unlike many machine learning models that feel like "black boxes," Decision Trees provide a clear illustration of how a decision was made.

### Step 1: Load the Data
We'll begin by importing a few libraries we'll need and loading the data (which is included with sci-kit-learn, a standard library for machine learning).

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn import tree

# set seed for reproducability
np.random.seed(52179)

# Load the dataset
cancer = load_breast_cancer()

# Print the available keys
print(cancer.keys())
```

These are the keys in the dataset that we can use. `data` contains the features, `target` contains the labels, `DESCR` is a full description of the dataset, `feature_names` are the names of the features, and `target_names` are the names of the different types of cancer.

### Step 2: Examine the Data
Next, let's look at the data and understand it.

```python
# Create a DataFrame
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)

# Add a target column to the DataFrame
df['target'] = cancer.target

# Display the first 5 rows of the DataFrame
df.head()
```

The DataFrame consists of several features that are characteristics of cell nuclei from a digitized image of a fine needle aspirate (FNA) of a breast mass. The 'target' column indicates whether the mass is malignant (0) or benign (1).

### Step 3: Create a Test-Train Split
Next, we'll split our data into **training data** (which we'll be using to train our decision tree) and **test data** (which we'll use to evaluate how well our decision tree does at distinguishing between cancerous and benign masses).

```python
# Define features and target
X = df[cancer.feature_names]
y = df['target']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

### Step 4: Train the Decision Tree
Now, we can finally create our decision tree!

```python
clf = DecisionTreeClassifier(
    max_depth=3,  # Limit tree depth to 3
    ccp_alpha=0.01  # Complexity parameter used for Minimal Cost-Complexity Pruning
)
clf.fit(X_train, y_train)
```

### Step 5: Interpret the Decision Tree
Now, let's "interpret" the decision tree by creating a visualization of it. At the "top" of the decision tree, you see ALL the cases in our data set bundled together (basically, this assumes you've done no tests or analysis at all). Then, at each level of the decision tree, we "split" on one of the features. For example:

1. The most important feature (at the very top level) is whether the "worst_perimeter" is less than 105. If it is, the tumor is likely to be malignant.
2. At the second level, we can further "subdivide" based on things like "worst concave points" or "worst feature". The further we go down in tree, the more "pure" nodes get (that is, each node will increasinly contain ONLY benign cases or cancerous cases, instead of a "mix").
3. The ultimate goal is to predict whether some new case is cancerous. More on that next!

```python
# Plot the Decision Tree
plt.figure(figsize=(15,10))
tree.plot_tree(clf, filled=True, feature_names=cancer.feature_names, class_names=cancer.target_names)
plt.show()
```

This tree can be seen as a flowchart that a clinician might use to determine the likelihood of cancer. Each decision point (node) considers a specific characteristic of a tumor, such as its size (worst perimeter), texture, and concavity (worst concave points). The "worst" prefix indicates that the measurement is the most extreme value observed for that feature. The **gini score** at each node is a measure of purity - a low gini score indicates that the samples at that node are very similar to each other in terms of the outcome (i.e., mostly benign or mostly malignant), while a higher score indicates a mixture of benign and malignant samples.

The samples is the number of observations that fall into each node, and value shows the distribution of benign and malignant cases at that node. Finally, the class is the predicted classification based on the data flowing into that node.

To use this tree for diagnosis:

1. You start at the top node and answer the question based on the tumor's characteristics.
2. Depending on your answer, you move down to the left or right branch.
3. This process continues until you reach a leaf node, which gives you the predicted classification based on the characteristics observed.
4. For example, if a patient's tumor has the worst concave points greater than 0.135, you follow the branch to the right and then consider the next characteristic. This step-by-step approach continues until a final prediction is made.

### Step 6: Evaluate the Model
Finally, we'll evaluate the model by using our test data. This will tell us the proportion of times we get things "right" (correctly classifying tumors as cancerous or benign) versus "wrong" (for example, classifying a tumor that is actuall cancerous as benign). As it turns out, we can do pretty good!

```python
# Predict the response for the test dataset
y_pred = clf.predict(X_test)

# Model Accuracy
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

This means the system can predict new cases of breast cancer with around 91% accuracy.

### What is the Turing Test and Why is it Important?

The **Turing Test**, proposed by Alan Turing in 1950, is a method for determining whether a machine can exhibit intelligent behavior equivalent to, or indistinguishable from, that of a human. Turing's proposal addressed the question: "Can machines think?" by suggesting a practical test for machine intelligence.

In the Turing Test, a human judge engages in a natural language conversation with one human and one machine, both of which are hidden from view. If the judge cannot consistently tell which is which, the machine is said to have passed the test. The significance of this test lies not in the machine's ability to give correct answers but in its capacity to exhibit human-like responses.
The Turing Test, formulated by Alan Turing in 1950, is a foundational concept in the field of Artificial Intelligence (AI), though it is not without its limitations and criticisms. To better appreciate this, we need to delve into both the essence of the Turing Test and Turing's broader ideas about computation and its limits.


The main criticism of the Turing Test lies in its emphasis on imitation over understanding. It doesn't necessarily measure a machine's comprehension or consciousness but rather its ability to mimic human responses. This limitation means that a machine could pass the Turing Test without possessing what we consider broader aspects of intelligence, such as emotional understanding, ethical reasoning, or creative thinking. There's also a philosophical argument that replicating human conversation is not synonymous with possessing a mind or consciousness.

Turing was a visionary in the field of computation, and his ideas extended far beyond the Turing Test. He was instrumental in the development of early computers and laid the groundwork for modern computational theory. Turing was aware of the limits of computation, as evident in his work on the **Halting Problem**, which proved that no algorithm can universally determine whether any given algorithm will eventually halt or run indefinitely. This work highlights a fundamental boundary of computational systems.

Modern AI, especially developments in chatbots and conversational AI, owes much to Turing's legacy. For instance, technologies like Google Bard, Character.AI, Maude, GPT-4, and others are designed to generate human-like text based on input. While systems like GPT-4 showcase remarkable advancements in AI, they also illustrate the limitations outlined by Turing's Test. These systems can generate text that is often indistinguishable from human writing in many contexts, but they do so based on patterns in data rather than genuine understanding or consciousness.

## Five Principles of AI Ethics
Ethical concerns in AI encompass a range of issues related to the development, deployment, and consequences of AI technologies. These concerns often reflect broader societal values and can significantly impact both individuals and communities.

While there isn't yet any universally accepted "standard" for AI ethics, a number of scholars have proposed adopting the "four principles" approach that is popular in medicine and biological reseach. Many suggest adding a "fifth" principle (of transparency).

**Respect for Autonomy** in AI is about safeguarding and promoting the agency of individuals. In practice, this means AI systems should be designed in such a way that they support human decision-making rather than replace it. For instance, a healthcare AI that suggests treatment options empowers doctors and patients to make informed decisions, rather than dictating a specific course of action. This principle also entails giving users clear options to opt-in or opt-out of AI-driven decisions. For example, a user of a social media platform should have the option to choose whether an AI curates their newsfeed or not, thus preserving their autonomy in the digital space.

**Non-Maleficence** requires that AI systems do no harm. This involves rigorous testing of AI systems for safety and reliability, especially in high-stakes domains like autonomous vehicles or medical diagnostics. An example here would be the implementation of fail-safes in autonomous cars to prevent accidents. Additionally, it means implementing measures to prevent misuse of AI that could lead to harm, such as using facial recognition technologies responsibly to avoid infringing on individuals' privacy and rights.

**Beneficence** mandates that AI should actively contribute to the well-being of individuals and society. A notable example is the use of AI in environmental science, such as analyzing satellite data to monitor climate change and help in disaster management. Here, the AI is not just benign but actively beneficial, aiding in understanding complex global phenomena that are crucial for human welfare. Another example is AI-assisted personalized education, where AI systems analyze students' learning patterns and customize educational content, thereby enhancing the learning experience and outcomes.

**Justice** as a principle in AI ethics stresses the fair and equitable distribution of both the benefits and burdens of AI. This means addressing biases in AI algorithms that could perpetuate social inequalities. An instance of this is the ongoing effort to eliminate racial or gender biases in AI recruitment tools, ensuring that job recommendations are fair and not skewed by underlying discriminatory algorithms. It also involves making sure that the benefits of AI, such as advanced healthcare diagnostics, are accessible to all segments of society, not just the privileged few.

**Transparency** in AI is key to building trust and ensuring accountability. This principle requires clear communication about how AI systems function, the kind of data they utilize, and the logic behind their decisions. For instance, when a credit scoring AI denies a loan application, the applicant should be able to understand the factors that led to this decision. Transparency also involves making the workings of AI systems understandable to non-experts, thus demystifying AI and fostering a more informed public discourse about its capabilities and limitations.

In sum, these principles serve as a moral compass in the rapidly evolving landscape of AI, ensuring that technological advancements are not at odds with human values and societal norms. They provide a robust ethical framework that can guide all stakeholders involved in AI development  from programmers and data scientists to policymakers and end-users  towards creating AI that is not only powerful and efficient but also respectful, safe, just, and transparent.
