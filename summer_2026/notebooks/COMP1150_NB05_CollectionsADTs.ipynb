## Lists: Ordered Things You Can Still Change

**Gwendolen Fairfax** runs Fairfax Stationery on Mayfair High Street. Customers ring through orders all day, and Gwendolen wants them recorded **in the order they arrived** — first come, first served, no exceptions. She also reserves the right to cross items off, add a forgotten one, or revise a misheard request.

That is a **list**: a row of items kept in order, with a position (an **index**) for each one, and free permission to change the contents later.

````python
# ⬇️ CHANGE THIS, THEN RE-RUN
todays_orders = ["monogrammed cards", "ink, midnight blue", "two journals", "wax seals"]
# ----------------------------------

print(todays_orders)
print("How many orders today?", len(todays_orders))
````

### Understanding the Code

- `todays_orders` is a **list** — the square brackets `[ ]` are how Python recognises one.
- The items keep the order they were written in. The first entry is at **index 0**, not 1. Gwendolen reads top-down; Python counts from zero.
- `len(todays_orders)` reports the length. It is the single most-used function on a list.

### 🔮 Predict Before You Run

The next cell prints three things: the item at a chosen index, the last item, and the first two items as a slice. Before pressing Run, **write down what each line will print** with `position = 0`. Then run and check. Disagreement is good — that's where the learning happens.

### Reading and Slicing

Gwendolen wants the very first order (the urgent one) and the last (the most recent). She can ask Python by **index**.

````python
# ⬇️ CHANGE THIS INDEX, THEN RE-RUN
position = 0
# ----------------------------------

print("Item at position", position, "→", todays_orders[position])
print("Last item       →", todays_orders[-1])
print("First two       →", todays_orders[:2])
````

### Understanding the Code

- `todays_orders[0]` is the first item. `todays_orders[-1]` is the last — negative indices count back from the end.
- `todays_orders[:2]` is a **slice**: a sub-list from the start up to (but not including) position 2. Slices give you a new list; the original is untouched.

### Changing the List

Lists are **mutable**: you can rewrite a slot, append to the end, or remove an item by value. Gwendolen does all three before lunch.

````python
orders = ["monogrammed cards", "ink, midnight blue", "two journals", "wax seals"]

orders.append("calligraphy pen")     # add to the end
orders[1] = "ink, oxblood"           # correct a misheard order
orders.remove("wax seals")           # the customer changed her mind

print(orders)
````

### Understanding the Code

- `append(x)` adds `x` to the end of the list. The list grows by one.
- `orders[1] = ...` overwrites position 1. Nothing is added; one entry is replaced.
- `remove(x)` deletes the **first** matching value. If the item isn't there, Python raises an error — Gwendolen would prefer to be told than to lose a sale silently.

**Takeaway:** reach for a list when the **order matters** and the contents will **change**.

## A Trap: Two Names for the Same List

Gwendolen writes her morning list, then makes a "copy" before tweaking it for her evening shift. Or so she thinks.

````python
morning = ["cards", "ink", "journals"]
evening = morning              # NOT a copy!

evening.append("wax seals")    # editing evening...

print("evening :", evening)
print("morning :", morning)    # ...but morning changed too!
````

### Understanding the Code

`evening = morning` does **not** make a new list. It makes a second **name** that points at the *same* list in memory. Editing through one name shows up through the other. Python calls this **aliasing**, and it catches every beginner.

The diagram below is the picture of what just happened — two arrows, one list.

````python
from graphviz import Digraph

a = Digraph(graph_attr={"rankdir": "LR"})
a.node("M", "morning", shape="ellipse", style="filled", fillcolor="lightyellow")
a.node("E", "evening", shape="ellipse", style="filled", fillcolor="lightyellow")
a.node("L", "['cards', 'ink', 'journals', 'wax seals']",
       shape="box", style="filled", fillcolor="lightblue")
a.edge("M", "L")
a.edge("E", "L")
a
````

**Reading it:** the two yellow ellipses are *names*; the blue box is the *list*. Both names point at the same list, so a change made through either name shows up through both.

### The Fix: Make a Real Copy

When Gwendolen genuinely wants an independent second list, she uses `.copy()` (or the full slice `morning[:]`, or `list(morning)` — all three do the same job).

````python
morning = ["cards", "ink", "journals"]
evening = morning.copy()       # NEW list with the same contents

evening.append("wax seals")

print("evening :", evening)
print("morning :", morning)    # untouched
````

### Understanding the Code

`morning.copy()` returns a fresh list with the same items inside. Now `morning` and `evening` are two different lists; editing one leaves the other alone.

**Takeaway:** mutability is power, but it is also danger. The tuples in the next section avoid this whole family of bugs by **refusing to be edited** — which is exactly why Augusta Bracknell prefers them for permanent records.

## Tuples: Fixed Records

**Augusta Bracknell** runs a vetting firm two doors down. Every applicant arrives as the same four pieces of information, in the same order, *every single time*: name, age, address, and (most importantly) annual income. She does not invite editing.

A **tuple** is a list's stricter cousin: ordered, indexable, and **immutable** — once built, it cannot be changed. That is not a limitation; it is a promise. When you see a tuple, you know the shape is fixed.

### 🔮 Predict Before You Run

In the next cell, `applicant` is a tuple of four items in the order **(name, age, address, income)**. After unpacking, what will the two `print` lines show? Predict, then run.

````python
# ⬇️ CHANGE THESE FIELDS, THEN RE-RUN
applicant = ("Jack Worthing", 29, "Belgrave Square", 7000)
# ----------------------------------

name, age, address, income = applicant      # unpacking

print(name, "is", age, "years old.")
print("Annual income:", income, "pounds.")
````

### Understanding the Code

- Round brackets `( )` and commas make a tuple. The four fields are positional: position 0 is the name, position 3 is the income — always, no exceptions.
- `name, age, address, income = applicant` is **tuple unpacking**: Python hands the four positions to four variables in one line. This is the cleanest way to read a fixed-shape record.
- Try `applicant[2] = "Half-Moon Street"`. Python will refuse. That refusal is the whole point.

### Returning Two Things at Once

A function can only return one value — but that one value can be a tuple, which is how Python returns "two things" without fuss. Augusta asks her vetting function for **both** the verdict and a reason.

````python
def vet(income):
    if income >= 5000:
        return ("approved", "income meets the threshold")
    else:
        return ("declined", "income below threshold")

verdict, reason = vet(7000)
print("Verdict:", verdict)
print("Reason :", reason)
````

### Understanding the Code

- The function returns **one tuple** containing two strings. Python lets us pretend it returned two things by unpacking on the receiving line.
- This pattern — verdict + reason, value + error, x + y — is how nearly all multi-return Python is written. You will see it again in every later notebook.

**Takeaway:** use a tuple when the record has a **fixed shape** and shouldn't be edited after it's built. Use a list when the contents will grow, shrink, or be rewritten.

## Dictionaries: Look It Up by Name

**Algernon Moncrieff** caters Mayfair's small parties. His menu has a dozen items, and at any moment a customer may ask the price of one. Algernon used to scan a printed list with his finger. Now he asks Python.

A **dictionary** stores **key–value pairs**. Give it the key, and it hands back the value in a single step — no scanning, no counting positions, no order to remember.

````python
# ⬇️ CHANGE THESE PRICES, THEN RE-RUN
menu = {
    "cucumber sandwich":  0.75,
    "muffin":             0.50,
    "tea, Earl Grey":     0.40,
    "champagne, glass":   3.00,
}
# ----------------------------------

print("A muffin costs:", menu["muffin"])
````

### Understanding the Code

- The curly braces `{ }` and the colons mark a **dictionary**.
- The string on the left of each colon is a **key**; the number on the right is its **value**.
- `menu["muffin"]` is a **key lookup**. Python goes straight to that entry. It does *not* walk through the others first.

### Adding, Updating, and Checking First

Algernon adds a new item and corrects a price using the same `key = value` syntax. He also checks whether something is on the menu *before* trying to read it — looking up a key that isn't there is an error.

````python
menu["scone"] = 0.60            # add a new key
menu["muffin"] = 0.55           # update an existing one

if "cucumber sandwich" in menu:
    print("Cucumber sandwich:", menu["cucumber sandwich"])

if "caviar" not in menu:
    print("No caviar on this menu — Algernon would never.")
````

### Understanding the Code

- Assigning to a new key **adds** it. Assigning to an existing key **overwrites** it. Same syntax, two outcomes — Python picks based on whether the key already exists.
- `key in some_dict` returns `True` or `False`. Use it whenever you aren't certain the key is there.

### Safer Lookup: `.get()`

`menu["caviar"]` crashes when caviar isn't on the menu. Often that is what Algernon wants — but sometimes "missing" should quietly mean *zero* or *not available*. The `.get()` method gives him that option.

````python
print(menu.get("scone"))              # found → returns 0.60
print(menu.get("caviar"))             # missing → returns None
print(menu.get("caviar", 0.00))       # missing → returns the default we supply
````

### Understanding the Code

- `d.get(key)` returns the value if the key exists, and **`None`** if it doesn't — *without* crashing. (`None` is Python's "nothing here" value.)
- `d.get(key, default)` returns *your* chosen default instead of `None`. Algernon uses `0.00` when missing means "not for sale today."
- **Rule of thumb:** use `.get()` when missing is a normal possibility. Use `d[key]` when missing should be a loud bug.

### Walking the Whole Menu

Sometimes Algernon does want every item — for example, to print the whole price list. The `.items()` view yields each key and its value together.

````python
for item, price in menu.items():
    print(f"{item:<22} £{price:.2f}")
````

### Understanding the Code

- `menu.items()` yields one `(key, value)` tuple per entry — note the tuples from the last section, quietly working underneath.
- The two variables `item, price` are tuple-unpacked on every pass of the loop. This is the standard way to iterate a dictionary.

### Sorting: From Most Expensive Down

Algernon prints a price list ordered from most to least expensive. The built-in `sorted()` returns a new sorted list — and it works on a dictionary's `.items()` view too.

````python
priciest = sorted(menu.items(), key=lambda kv: kv[1], reverse=True)

for item, price in priciest:
    print(f"{item:<22} £{price:.2f}")
````

### Understanding the Code

- `sorted(some_sequence)` returns a **new** sorted list. The original is untouched. (There is also `list.sort()`, which sorts a list **in place** and returns nothing.)
- `key=lambda kv: kv[1]` tells `sorted` *what* to sort by. Each `kv` is a `(name, price)` tuple, and `kv[1]` is the price. Read it as: *"sort using the second piece of each pair."*
- `reverse=True` flips the order so most-expensive comes first.

You will meet `key=` again any time you sort something more interesting than plain numbers. The cheatsheet near the end of this notebook collects the other variants worth knowing about.

**Takeaway:** reach for a dictionary when you want to **look something up by name** — by a product, a person, an ID. Lists are for *positions*; dictionaries are for *labels*.

## Sets: Membership, Overlap, Uniqueness

**Cecily Cardew** runs a small country school. Today she has two cohorts — the morning Latin class and the afternoon French class — and one urgent question from the bursar: **which students are in *both*?**

A **set** is a collection with two strict rules: **no duplicates**, and **no order**. What sets are extraordinarily good at is comparing memberships — who's in A, who's in B, who's in both, who's only in one.

````python
# ⬇️ CHANGE THESE ROSTERS, THEN RE-RUN
latin   = {"Ada", "Beatrice", "Cyril", "Dorian", "Esme"}
french  = {"Beatrice", "Cyril", "Felix", "Gwendolen"}
# ----------------------------------

print("In both classes :", latin & french)        # intersection
print("In either class :", latin | french)        # union
print("Latin only      :", latin - french)        # difference
print("Exactly one     :", latin ^ french)        # symmetric difference
````

### Understanding the Code

- Curly braces with **no colons** make a set, not a dictionary. (An empty set is `set()`; `{}` is an empty dict — Python's one truly awkward corner.)
- `&` is **intersection** (in both), `|` is **union** (in either), `-` is **difference** (in the first but not the second), `^` is **symmetric difference** (in exactly one).
- Cecily's answer to the bursar is one line: `latin & french`.

### 🔮 Predict Before You Run

The next cell builds a set from a messy list with six entries — but only four distinct names. **Predict** two things before you run it: (1) what will `len(guests_unique)` print? (2) Will the unique guests print in the same order as the original list, or not?

````python
guests_messy = ["Ada", "Beatrice", "Ada", "Cyril", "Beatrice", "Dorian"]
guests_unique = set(guests_messy)

print("Original :", guests_messy)
print("Unique   :", guests_unique)
print("How many distinct guests?", len(guests_unique))
````

### Understanding the Code

- `set(some_list)` builds a set from any sequence and drops every repeat.
- The result has **no defined order** — Python may print the names in any order. If Cecily wants a sorted list back, she calls `sorted(guests_unique)`.
- Membership in a set is also a single-step check: `"Ada" in guests_unique` is fast no matter how large the set grows.

**Takeaway:** use a set when you care about **whether things are present**, not where. Use it when you need **uniqueness**.

### Choosing Your Collection — A Quick Guide

Four collections, four jobs. If you remember nothing else from this notebook, remember this table.

| If the job needs… | …reach for | Example from this notebook |
|---|---|---|
| Order *and* the ability to add / remove / edit | **list** | Gwendolen's daily orders |
| A fixed-shape record that shouldn't change | **tuple** | Augusta's applicant `(name, age, address, income)` |
| Looking up a value by a label / name / ID | **dict** | Algernon's menu (item → price) |
| Membership / uniqueness / overlap | **set** | Cecily's "in both classes" question |

If you find yourself writing a `for` loop just to check whether something is in a list, that's a **set** screaming. If you keep using `if` to find a price, that's a **dict**.

## Nested Collections: Real Data Has Shape

So far each example has been **flat** — one list, or one dictionary, or one set. Real data is almost never that simple. Algernon caters a dinner party tonight, and every guest has a **name**, an **age**, *and* a **dietary restriction**. That is three facts per person — one row, three columns.

The standard Python answer is a **list of dictionaries**: a list (one entry per guest) where each entry is a dict (one key per fact). This is exactly the shape of nearly every API response, every spreadsheet row, every database query you will meet in NB09 and NB10.

````python
party = [
    {"name": "Augusta",  "age": 60, "diet": "no shellfish"},
    {"name": "Cecily",   "age": 18, "diet": "vegetarian"},
    {"name": "Algernon", "age": 27, "diet": "none"},
    {"name": "Jack",     "age": 29, "diet": "no nuts"},
]

print(party[0])                 # a whole guest record
print(party[0]["name"])         # one field of one guest
````

### Understanding the Code

- `party` is a list. Each item in that list is a dictionary.
- `party[0]` returns the *whole first guest* — a dictionary. To drill into one field, you index **again**: `party[0]["name"]`.
- This double-step — *list index, then dict key* — is the rhythm of working with nested data. Get used to it; it is everywhere.

### Walking and Filtering the Party

A `for` loop over the list gives Algernon one full guest record at a time. Inside the loop, each `guest` is a dictionary — so guest fields use the dict syntax `guest["name"]`.

````python
for guest in party:
    print(f"{guest['name']:<10} (age {guest['age']:>2}) — {guest['diet']}")

# Filter: who is vegetarian?
vegetarians = [g["name"] for g in party if g["diet"] == "vegetarian"]
print()
print("Vegetarians:", vegetarians)
````

### Understanding the Code

- Inside the loop, `guest` is one whole dictionary. We pull each field with `guest["age"]`, `guest["diet"]`, etc.
- The last line is a **list comprehension** with a filter: `[expression for item in source if condition]`. Read it as *"give me the names, for every guest, whose diet is vegetarian."* You will meet comprehensions again in the ADT section.

**Takeaway:** flat collections are the stepping stones. **Lists of dicts** are the workhorse shape for real data. When NB09 returns rows from a database and NB10 returns JSON from an API, this is what they will look like.

## Abstract Data Types: The Promise vs. the Storage

**Jack Worthing** runs a concierge service. Customers form a line; he serves them in the order they arrived. His staff don't care *how* the line is stored — paper ledger, whiteboard, app — they only care that the system answers three questions consistently:

- **add** a customer to the back of the line
- **next_customer** — remove and return the one at the front
- **size** — how many are waiting

That trio is the **interface**: the promise the line makes. Jack could store the line as a Python list, or as a dictionary keyed by ticket number — the interface stays the same. That separation is what we call an **abstract data type**, or **ADT**.

> **An ADT is *what* the collection does. Its implementation is *how* it stores the data.**

The big idea: the rest of your program should only ever talk to the interface. Swap the storage out later and nothing else breaks.

### The Same Methods, in Pseudocode First

Before any Python, Jack writes the three methods in pseudocode — exactly the workflow from Notebook 3.

```text
CLASS CustomerQueue
  ON CREATE:
    line ← empty list

  METHOD add(name):
    APPEND name to line

  METHOD next_customer():
    REMOVE and RETURN the first item of line

  METHOD size():
    RETURN the length of line
```

Each method is one tiny operation on `line`. The Python below is the same pseudocode in working code.

````python
class CustomerQueue:
    def __init__(self):
        self._line = []                      # storage — a list

    def add(self, name):
        self._line.append(name)

    def next_customer(self):
        return self._line.pop(0)             # remove + return the front

    def size(self):
        return len(self._line)

q = CustomerQueue()
q.add("Gwendolen"); q.add("Algernon"); q.add("Cecily")

print("Waiting:", q.size())
print("Now serving:", q.next_customer())
print("Still waiting:", q.size())
````

### Understanding the Code

- The list `self._line` is the **implementation**. The leading underscore is a convention meaning *don't touch this from outside*.
- `add`, `next_customer`, and `size` are the **interface**. Jack's staff call only these.
- If next week we decided to store the queue as a dictionary keyed by arrival time, every line outside this class could stay exactly as it is.

### Picture It: One Interface, Two Implementations

The diagram below shows the same `CustomerQueue` interface sitting above two different storage choices. The point isn't the diagram — it's that the box on top doesn't care which box on the bottom you pick.

````python
from graphviz import Digraph

d = Digraph(graph_attr={"rankdir": "TB"})
d.node("I", "CustomerQueue (interface)\nadd · next_customer · size",
       shape="box", style="filled,bold", fillcolor="lightyellow")
d.node("L", "list-backed\nself._line = []", shape="box",
       style="filled,dashed", fillcolor="lightgrey")
d.node("D", "dict-backed\nself._line = {ticket: name}", shape="box",
       style="filled,dashed", fillcolor="lightgrey")
d.edge("I", "L", label="one option")
d.edge("I", "D", label="another option")
d
````

**Reading it:** the yellow box is the public promise — the three method names. The grey boxes are two different ways to keep the data. Code that uses the queue only touches the yellow box, so changing the grey box is free.

### A Cleaner Way to Build a Collection: Comprehensions

**Arthur Goring**, the bespoke tailor next door, has a list of inseam measurements and a small rule from his master cutter: add 2 cm to every measurement for the seam allowance. He *could* write a loop. He prefers a **list comprehension** — one line that builds a new list from an old one.

````python
inseams = [78, 81, 75, 84, 79]

# Old way: loop + append
adjusted_loop = []
for m in inseams:
    adjusted_loop.append(m + 2)

# Comprehension way: one line
adjusted = [m + 2 for m in inseams]

print("loop way        :", adjusted_loop)
print("comprehension   :", adjusted)

# A dict comprehension — same idea, key:value shape
labelled = {f"client_{i}": m + 2 for i, m in enumerate(inseams, start=1)}
print("labelled        :", labelled)
````

### Understanding the Code

- A **list comprehension** has the shape `[expression for item in source]`. Read it left-to-right: *"build a list of expression, for each item in source."*
- A **dict comprehension** is the same idea with `key: value` on the left and curly braces around the whole thing.
- Both are equivalent to a loop with `append`/assignment — Arthur writes them this way because they read closer to what he means: *"two more centimetres on everything."*

**Takeaway:** ADTs are about **promises**. Comprehensions are about **expressing the promise concisely**. Both move you from *"I'm pushing data around"* toward *"I'm describing what I want."* 

### One Last Picture: Why the Right Collection Matters

Algernon's menu has hundreds of items some days. Searching a list means checking each entry until you find a match; a dictionary jumps straight there. The chart below times both as the menu grows.

````python
import time
import matplotlib.pyplot as plt

sizes = [100, 1000, 10000, 100000]
list_times, dict_times = [], []

for n in sizes:
    items   = [f"item_{i}" for i in range(n)]
    as_list = items
    as_dict = {k: 1 for k in items}
    target  = f"item_{n - 1}"             # worst case: at the end

    t0 = time.perf_counter()
    _  = target in as_list                 # scans
    list_times.append(time.perf_counter() - t0)

    t0 = time.perf_counter()
    _  = target in as_dict                 # one step
    dict_times.append(time.perf_counter() - t0)

plt.plot(sizes, list_times, marker="o", label="list (scan)")
plt.plot(sizes, dict_times, marker="s", label="dict (lookup)")
plt.xscale("log"); plt.yscale("log")
plt.xlabel("number of items"); plt.ylabel("seconds to find one")
plt.title("Algernon's menu: list vs. dict lookup")
plt.legend(); plt.grid(True, which="both", linestyle=":"); plt.show()
````

**Reading it:** both axes are log-scaled, so a straight diagonal means *"grows with the data"* and a flat line means *"doesn't care how big the data is."* The list line climbs. The dict line barely moves. That gap is why we picked a dict for the menu and not a list — and it's exactly the kind of trade-off Notebook 7 will study under the name **Big O**.

## A Methods Cheatsheet

The methods you used in this notebook, plus a few close cousins worth knowing exist. You don't need to memorise the right-hand column — just remember that this table is here.

### Lists

| Method | What it does |
|---|---|
| `lst.append(x)` | Add `x` to the end |
| `lst.remove(x)` | Remove first matching value (errors if absent) |
| `lst.pop(i)` | Remove and return item at index `i` (default: the last) |
| `lst.insert(i, x)` | Insert `x` at position `i` |
| `lst.extend(other)` | Add every item of `other` to the end |
| `lst.copy()` | Return a new, independent list (the fix for aliasing) |
| `lst.sort(key=..., reverse=...)` | Sort **in place** — returns nothing |
| `sorted(lst, key=..., reverse=...)` | Return a **new** sorted list — leaves `lst` alone |
| `len(lst)` &nbsp; • &nbsp; `x in lst` | Length • membership (scans) |

### Dictionaries

| Method | What it does |
|---|---|
| `d[key]` | Look up (errors if missing) |
| `d.get(key, default)` | Look up with a fallback — never crashes |
| `d[key] = v` | Add a new key or overwrite an existing one |
| `d.pop(key)` | Remove and return the value |
| `d.items()` • `d.keys()` • `d.values()` | The three views you can iterate |
| `key in d` | Membership test (fast) |

### Sets

| Method / operator | What it does |
|---|---|
| `s.add(x)` | Add `x` (no-op if already in) |
| `s.discard(x)` | Remove `x` (no error if absent) |
| `a & b` &nbsp; • &nbsp; `a \| b` | Intersection • union |
| `a - b` &nbsp; • &nbsp; `a ^ b` | Difference • symmetric difference |
| `x in s` | Membership test (fast) |

## Key Terms

- **Abstract Data Type (ADT)** — A description of *what* a collection does (its methods and their behaviour) separate from *how* it stores the data.
- **Aliasing** — When two variables point at the same underlying object, so a change made through one is visible through the other.
- **Comprehension** — A compact one-line way to build a new list or dictionary from an existing sequence.
- **Copy** — A new, independent object with the same contents. For a list, `lst.copy()`, `list(lst)`, or `lst[:]`.
- **Default value** — The value returned by `d.get(key, default)` when the key is missing.
- **Dictionary** — A collection of key–value pairs that supports fast lookup by key. Written with `{key: value, ...}`.
- **Difference (sets)** — The elements in one set but not another. Written `a - b`.
- **Immutable** — Cannot be changed after it is created. Tuples are immutable; lists are not.
- **Implementation** — The actual storage and code that fulfils an ADT's promise.
- **Index** — The position of an item in an ordered collection. The first index in Python is `0`.
- **Interface** — The methods an ADT promises to support. Other code talks only to the interface.
- **Intersection (sets)** — The elements common to two sets. Written `a & b`.
- **Key** — The label used to look up a value in a dictionary. Must be unique within that dictionary.
- **List** — An ordered, changeable collection of items. Written with `[ ]`.
- **List of dictionaries** — The standard shape for tabular real-world data: one dict per row, one key per column.
- **Mutable** — Can be changed after it is created. Lists, dicts, and sets are mutable.
- **`None`** — Python's built-in "nothing here" value. Returned by `d.get(key)` when the key is missing.
- **Set** — An unordered collection with no duplicates. Written with `{ }` (no colons).
- **Slice** — A sub-range of a list, written `list[start:stop]`, producing a new list.
- **Sort key** — The function passed to `sorted(..., key=...)` that tells Python *what* to sort by.
- **Tuple** — An ordered, **immutable** collection. Written with `( )`.
- **Union (sets)** — All elements that appear in either set. Written `a | b`.
- **Unpacking** — Assigning the contents of a tuple (or list) to several variables in one statement.
- **Value** — The data stored under a key in a dictionary. Any type is allowed.

## Summary

- **Pick by promise.** Lists are *ordered + changeable*. Tuples are *ordered + fixed*. Dictionaries are *labelled lookup*. Sets are *membership + uniqueness*. The decision table is the cheatsheet.
- **Watch for aliasing.** `b = a` makes a second name for the same list, not a copy. `a.copy()` (or a full slice) is the fix.
- **Real data is nested.** A *list of dictionaries* is the everyday shape — one dict per row, one key per field.
- **Lookup style matters.** Walking a list grows slower as the list grows; a dictionary lookup stays fast. The seed of next-up Big O.
- **ADTs separate promise from storage.** Hide your storage behind methods and the rest of your program survives a redesign.
- **Comprehensions** turn a build-a-collection loop into one readable line.
