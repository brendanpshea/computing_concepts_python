# Plain Text Files (txt), ASCII, and Unicode
**Plain text files** are among the simplest kinds of computer files. They contain only text characters, without any special formatting or metadata (like bold, italics, or hyperlinks). The most common plain text file format has the extension '.txt'.

Plain text files are incredibly versatile. You can use them to jot down notes, write code, or even store numerical data. But how does your computer transform a sequence of 1s and 0s into meaningful text? The magic happens through character encoding standards like ASCII and Unicode.

### ASCII: The Basic Building Block

ASCII (American Standard Code for Information Interchange) was one of the first widely accepted standards for translating numbers into text. It uses seven bits to represent each character. Because each bit can be either a 0 or a 1, seven bits can represent 2^7 (128) different values.   Here's a tiny slice of the ASCII table to give you an idea:

| Character | Binary | Decimal | Hexadecimal |
| --- | --- | --- | --- |
| 'A' | 01000001 | 65 | 41 |
| 'a' | 01100001 | 97 | 61 |
| '3' | 01100011 | 51 | 33 |
| '+' | 00101011 | 43 | 2B |
| '@' | 01000000 | 64 | 40 |
| '$' | 00100100 | 36 | 24 |
| '%' | 00100101 | 37 | 25 |
| '?' | 00111111 | 63 | 3F |
| '(' | 00101000 | 40 | 28 |
| ')' | 00101001 | 41 | 29 |

Binary is the numeric representation of the character in binary, decimal is the same value in decimal, and character is the equivalent character.

ASCII can encode 128 different characters in total: 33 control characters (such as carriage return, line feed, and tab), and 95 printable characters including digits (0-9), uppercase and lowercase letters (A-Z, a-z), and special symbols (like !, @, #, $, %, &, *, (, ), and ?).



### Unicode: A more Universal Approach

As computers spread globally, it quickly became apparent that ASCII wasn't enough. After all, there are many more than 128 characters used in written languages around the world! To address the limitations of ASCII and provide a more comprehensive character encoding system, **Unicode** was developed. Unicode aims to provide a unique code point for every character across all writing systems, including various scripts, symbols, and emoji.

Unicode encompasses a wide range of character sets, including:

-   Basic Latin (ASCII)
-   Latin-1 Supplement (additional characters for Western European languages)
-   Cyrillic (for Russian and other Slavic languages)
-   Chinese, Japanese, and Korean (CJK) ideographs
-   Emoji and pictographs
-   Mathematical and scientific symbols

Unicode uses a variable-length encoding system, allowing characters to be represented using 8, 16, or 32 bits. The most common Unicode encoding is UTF-8 (Unicode Transformation Format 8-bit), which is backward-compatible with ASCII. In UTF-8, characters in the ASCII range are represented using a single byte, while characters beyond ASCII use two or more bytes.

The flexibility and extensibility of Unicode have made it the dominant character encoding standard in modern computing. It enables the reliable storage and exchange of text across different languages, platforms, and devices.

### Choosing the Right Encoding

When working with text files, it's essential to choose the appropriate character encoding. If the wrong encoding is used, characters may appear garbled or be misinterpreted. Some common scenarios to consider:

-   For English-only text or simple data interchange, ASCII is often sufficient.
-   For multilingual text or data that includes special characters and symbols, Unicode (UTF-8) is the preferred choice.
-   When working with languages that require a specific encoding (e.g., UTF-16 for some Asian languages), ensure that the encoding is explicitly specified and handled correctly.

Most modern programming languages and text editors support both ASCII and Unicode, making it easier to work with text files in various encodings. By understanding the differences between ASCII and Unicode and choosing the appropriate encoding for your needs, you can ensure that your text is stored and processed reliably across different systems and applications.

## Making it Pretty With Markup (html, md, xml)
While plain text is straightforward and simple, it's also... well, plain. If we want text that's bold or italic, or if we want to create headers, lists, or links, plain text falls short. That's where markup languages come in.

A **markup language** is a system for annotating a document in a way that is syntactically distinguishable from the text. In other words, a markup language lets you write text and also write instructions about how that text should be displayed or processed.  The term 'markup' comes from the traditional publishing practice of 'marking up' a manuscript, which involves adding symbolic printer's instructions in the margins.

Some of the most common markup languages you'll come across in computing include HTML, XML, and Markdown. Each of these serves a different purpose and uses a different set of tags to denote different types of content.



### HTML: The Language of the Web

HTML, which stands for Hypertext Markup Language, is the backbone of the World Wide Web. It is used to create the structure and content of web pages. HTML uses tags to define the elements of a document, such as headings, paragraphs, lists, links, images, and more.

Here's a simple example of HTML code:

```html
<!DOCTYPE html>
<html>
  <head>
    <title>My Web Page</title>
  </head>
  <body>
    <h1>Welcome to my website!</h1>
    <p>This is a paragraph of text.</p>
    <ul>
      <li>Item 1</li>
      <li>Item 2</li>
      <li>Item 3</li>
    </ul>
  </body>
</html>
```
HTML tags are enclosed in angle brackets (`<` and `>`), and most tags have an opening and closing tag. The content between the tags is affected by the specific tag's functionality. HTML documents are rendered by web browsers, which interpret the tags and display the content accordingly.

### XML: Extensible Markup Language

XML is often used for data exchange between different systems, configuration files, and storing structured data. It provides a flexible and self-describing format for representing information.

## Source Code Files: C, Java, and Python

Source code files are plain text files that contain the instructions and statements written in a programming language. These files are human-readable and can be edited using text editors or integrated development environments (IDEs). In this section, we will explore source code files for three popular programming languages: C, Java, and Python.

### C Source Code Files
C is a general-purpose, procedural programming language known for its efficiency and low-level control. C source code files typically have the file extension `.c`. Here's an example of a simple C program:

```c
#include  <stdio.h>
int  main()  {   
    printf("Hello, World!\n");   
    return  0;  
    }
```

In this example, the `#include` directive is used to include the standard input/output library `stdio.h`. The `main` function is the entry point of the program, and it contains a single statement that prints "Hello, World!" to the console using the `printf` function.

C source code files are compiled using a C compiler, such as GCC (GNU Compiler Collection) or Clang, which translates the code into machine-executable binary.

### Java Source Code Files
Java is an object-oriented programming language known for its portability and extensive standard library. Java source code files have the file extension `.java`. Here's an example of a simple Java program:

```java
public  class  HelloWorld  {   
  public  static  void  main(String[] args)  {   
    System.out.println("Hello, World!");   
    }  
  }
```

In Java, the source code is organized into classes. In this example, the `HelloWorld` class contains the `main` method, which is the entry point of the program. The `System.out.println` statement is used to print "Hello, World!" to the console.

Java source code files are compiled into bytecode (`.class` files) using the Java compiler (`javac`). The bytecode is then executed by the Java Virtual Machine (JVM), which provides portability across different platforms.

## Assembly Code and Machine Code: Bridging the Gap

When we write source code in high-level programming languages like C, Java, or Python, it needs to be translated into a form that the computer's processor can directly understand and execute. This is where assembly code and machine code come into play. In this section, we will explore assembly code, machine code, and the process of converting source code into machine code.

### Assembly Code
Assembly code, also known as **assembly language**, is a low-level programming language that is specific to a particular processor architecture. It represents a human-readable form of machine code and provides a more direct way to interact with the processor's instructions and registers.

Here's an example of assembly code for the x86 architecture:

```assembly
section .text
    global _start

_start:
    mov eax, 4            ; sys_write system call
    mov ebx, 1            ; stdout file descriptor
    mov ecx, message      ; address of the string to write
    mov edx, message_len  ; length of the string
    int 0x80              ; invoke the system call

    mov eax, 1            ; sys_exit system call
    xor ebx, ebx          ; exit status 0
    int 0x80              ; invoke the system call

section .data
    message db 'Hello, World!', 0xa
    message_len equ $ - message
```

In this example, the assembly code is divided into sections: `.text` for executable instructions and `.data` for defining data. The code uses specific instructions like `mov` to move data into registers, `int` to invoke system calls, and `xor` for bitwise operations.

Assembly code provides more control over the processor's operations and can be used for low-level programming, optimization, and system-level tasks. However, writing assembly code is more challenging and less portable compared to high-level languages.

### Machine Code

**Machine code**, also known as machine language or object code, is the lowest-level representation of a program that can be directly executed by a computer's processor. It consists of binary instructions that correspond to specific operations the processor can perform.

Machine code is specific to a particular processor architecture and is not human-readable. It is typically represented in hexadecimal or binary format.

Here's an example of machine code for the x86 architecture:

```
B8 04 00 00 00   ; mov eax, 4 BB 01 00 00 00   ; mov ebx, 1
B9 00 00 00 00   ; mov ecx, message
BA 0E 00 00 00   ; mov edx, message_len
CD 80            ; int 0x80

B8 01 00 00 00   ; mov eax, 1
31 DB            ; xor ebx, ebx
CD 80            ; int 0x80
```

Each line represents a machine instruction in hexadecimal format (remember, for the machine, this will be binary!). The processor fetches these instructions from memory and executes them directly.

### From Source Code to Machine Code

To convert source code into machine code, the following steps are typically involved:

1.  **Compilation.** The source code is compiled using a compiler specific to the programming language. The compiler translates the high-level source code into assembly code or machine code.
2.  **Assembly.** If the compiler generates assembly code, an assembler is used to translate the assembly code into machine code.
3.  **Linking.** If the program consists of multiple source files or relies on external libraries, a linker is used to combine the machine code from different files and resolve external references.
4.  **Loading.** The resulting machine code is loaded into memory by the operating system, and the program is executed by the processor.

Some programming languages, like C and C++, are typically compiled directly into machine code. Others, like Java, are compiled into bytecode (an intermediate representation), which is then executed by a virtual machine (the "Java Virtual Machine"). Interpreted languages, like Python, execute the source code directly (using the Python Interpreter) without explicit compilation. This is only possible because the JVM and Python interpreter are themselves machine programs (which were originally written in C and compiled), howefver.

## How Do Computers Store Data?

When we use computers to store and manage information, it's essential to understand how data is organized and stored. In this section, we will explore two common approaches to storing data: text-based formats and binary formats.

### Text-Based Formats
Text-based formats store data as plain text, using characters that are human-readable and easily editable. Two popular text-based formats are CSV and JSON.

**Comma-Separated Values (CSV)** is a simple and widely used format for storing tabular data. In a CSV file, each line represents a record or row, and each value within a row is separated by a comma (or sometimes another delimiter like a semicolon or tab).  Here's an example of storing member information for a computer science club in CSV format:

```
ID,Name,Email,Grade
1,Alex,alice@example.com,10
2,Betty,bob@example.com,11
3,Charlie,charlie@example.com,12
```

In this example, the first line contains the headers or column names, and each subsequent line represents a member's data. CSV files are easy to create, read, and parse using various programming languages and tools.

**Javascript Object Notation (JSON)** is a lightweight and popular format for storing and exchanging structured data. It uses a human-readable text format that consists of key-value pairs and arrays.  Here's an example:

```javascript

[   
  {   "ID":  1,   "Name":  "Alex",   "Email":  "alice@example.com",   "Grade":  10   },   
  {   "ID":  2,   "Name":  "Betty",   "Email":  "bob@example.com",   "Grade":  11   },   
  {   "ID":  3,   "Name":  "Charlie",   "Email":  "charlie@example.com",   "Grade":  12   }  
  ]
```

In this example, the data is represented as an array of objects. Each object represents a member and contains key-value pairs for properties like ID, Name, Email, and Grade. JSON is widely supported by programming languages and is commonly used for data interchange between systems.

## Relational Databases
Unlike CSV and JSON, which are text-based formats, **relational databases** store data in a binary, non-human-readable format, using a sequence of bytes. They are used when efficiency, performance, and data integrity are crucial.

A relational database organizes data into tables, where each table consists of rows (records) and columns (fields). Tables are related to each other based on common attributes or keys.

Here's an example of storing member and project information for a computer science club in a relational database:

Members Table:

| ID | Name | Email | Grade |
| --- | --- | --- | --- |
| 1 | Alex | <alex@example.com> | 10 |
| 2 | Betty | <betty@example.com> | 11 |
| 3 | Charlie | <charlie@example.com> | 12 |


Projects Table:

| ID | Title | Description | MemberID |
| --- | --- | --- | --- |
| 1 | Web App | School website | 1 |
| 2 | Game | Python game | 2 |
| 3 | Mobile App | Club management app | 3 |

In this example, we have two tables: "Members" and "Projects." The "Members" table stores information about each member, while the "Projects" table stores details about the projects. The "MemberID" column in the "Projects" table acts as a foreign key, establishing a relationship between the two tables.

Relational databases use a binary format to store data efficiently on disk. They provide features like data integrity, querying, indexing, and transactions. Popular relational database management systems (RDBMS) include MySQL, PostgreSQL, and SQLite.

On the other hand, binary formats like relational databases offer more advanced features and are designed for efficient storage, retrieval, and management of structured data. They provide data integrity, querying capabilities, and support for complex relationships between data entities.

Understanding how computers store data using different formats is crucial for designing and working with data-driven applications. By choosing the appropriate format based on the requirements of your project, you can effectively store, retrieve, and manipulate data in a computer science context.

## Picture This: Image Data in a Digital World

To a computer, an image is simply an array of numbers, each representing a color. But to us, it forms a picturesque scene, a memory, or a piece of art. How does this transformation happen? Let's dive into the world of image data.

### RGB Values: The Primary Colors of the Digital World

Images are comprised of tiny individual units called pixels, each pixel corresponding to a specific color. This color is often represented as an **RGB value**, an acronym for Red, Green, Blue.  RGB values are tuples of three numbers, each ranging from 0 to 255. For example, the RGB value of (255, 0, 0) represents pure red, (0, 255, 0) is pure green, and (0, 0, 255) is pure blue. By combining these primary colors in varying amounts, we can represent millions of different colors. The RGB value (0, 0, 0) represents black, while (255, 255, 255) represents white.

### Bitmap (BMP) Files: A Pixel-by-Pixel Story

The **Bitmap (BMP)** format is one of the simplest image file formats. It was developed by Microsoft for use on personal computers. The BMP format stores color information for each pixel in the image without any compression. This means that the file size of a BMP image can be quite large, but it also means that the image quality is high, as no information is lost.

For a simplistic understanding, imagine a 3x3 pixel image with RGB values as follows:

````python
import matplotlib.pyplot as plt
import numpy as np

# A 3x3 array for colored image
# each set of three numbers represents red, green, and blue
color_image = np.array([[[255, 0, 0],   [0, 255, 0],   [0, 0, 255]],
                        [[255, 255, 0], [0, 255, 255], [255, 0, 255]],
                        [[0, 0, 0],     [255, 255, 255],[127, 127, 127]]], dtype=np.uint8)

# Display the image
plt.imshow(color_image)
plt.axis('off') # to hide the axes
plt.show()
````

### JPEG Files: Picture Perfect Compression

**JPEG**, standing for Joint Photographic Experts Group (the team that developed the format), is a commonly used method of lossy compression for digital images. If you've ever taken a photo with a digital camera or a smartphone, you've likely used the JPEG format.

Unlike BMPs, JPEGs undergo a process of compression where some data is discarded to reduce the file size. This process involves a complex transformation of the image data into a format that can be stored more compactly, at the expense of some loss of image quality.

For example, imagine that same 3x3 pixel image we used before. In the BMP version, we store the exact RGB value of each pixel, which can take a lot of space for large images. In the JPEG version, the image might be transformed and compressed so that the color information is stored in a way that requires less space.

The amount of compression in a JPEG file can be adjusted. Higher levels of compression result in lower quality images but smaller file sizes, while lower levels of compression result in higher quality images but larger file sizes.

The JPEG format is incredibly popular for digital photography and online images due to its balance of quality and file size. Understanding how it works is a perfect example of how clever algorithms and understanding of human perception can lead to powerful tools in computer science.

````python
# Example showing the difference between BMP and JPEG
# Requires running previous cell
from PIL import Image
import os
import numpy as np


np.random.seed(52179)  # for reproducibility

def print_file_info(filename):
    file_size = os.path.getsize(filename)
    print(f"File: {filename}, Size: {file_size} bytes")

# Create a 100x100 array with random shades of blue
shades_of_blue = np.zeros((100, 100, 3), dtype=np.uint8)
shades_of_blue[:, :, 2] = np.random.randint(0, 256, size=(100, 100))  # Random blue values

# Create a PIL Image object from the NumPy array
img = Image.fromarray(shades_of_blue)

# Save the image as BMP and JPEG
img.save('shades_of_blue.bmp')
img.save('shades_of_blue.jpg', 'JPEG')

# Display file information
print_file_info('shades_of_blue.bmp')
print_file_info('shades_of_blue.jpg')
````

You'll notice that the JPEG is MUCH SMALLER than the BMP. However, given the fine level of detail, I bet you can also notice some "blurriness" introduced by the compression.

````python
import matplotlib.pyplot as plt

# Load the images
img_bmp = Image.open('shades_of_blue.bmp')
img_jpg = Image.open('shades_of_blue.jpg')

# Create a figure with two subplots
fig, axs = plt.subplots(1, 2)

# Display the BMP image in the first subplot
axs[0].imshow(img_bmp)
axs[0].axis('off')
axs[0].set_title('BMP')

# Display the JPEG image in the second subplot
axs[1].imshow(img_jpg)
axs[1].axis('off')
axs[1].set_title('JPEG')

# Show the figure
plt.show()
````

## Listening to the Bytes: Audio Files

Music, podcasts, voice chats, or your favorite YouTube videos - audio is a fundamental part of our digital experience. But how do our devices transform sequences of numbers into something that makes sense to our ears? Let's dive into the symphony of audio file formats, with a special focus on WAV, MP3, and the magic behind streaming audio.

### WAV Files: The Unaltered Symphony

Imagine you're attending a live concert. The musicians strike their instruments and the sound waves begin their journey towards your ears. If we could capture these sounds in "freeze frames" of time, each snapshot would show a complex pattern of waves. This process forms the essence of digital audio.

In a WAV file (Waveform Audio File Format), these audio 'snapshots' are stored in their raw, uncompressed form. Each of these snapshots, known as 'samples', records the strength (amplitude) of the sound wave at a specific moment in time. When these samples are played back rapidly - around 44,100 times a second for CD quality sound - your brain perceives it as continuous sound, almost indistinguishable from the live performance!

However, this high fidelity comes with a cost: file size. Since WAV files store all these samples without any compression, they can take up quite a bit of space. For instance, just a minute of CD-quality stereo sound can require more than 10 MB!

### MP3 Files: Clever Compression

What if you want to share your concert recording with friends over the internet? Uploading or downloading a full 60-minute WAV file would mean transferring over 600 MB, which could be time-consuming or even impossible with slower connections. That's when the MP3 format, a superstar of compressed audio, comes to our rescue.

MP3 files apply a smart trick - they eliminate the parts of the audio that our ears typically can't hear. This process, known as 'lossy compression', substantially shrinks the file size while keeping the perceived audio quality high. The same 60-minute recording that was 600 MB in WAV format might only be about 60 MB as an MP3!

However, it's important to note that some audio information is lost during this compression, so the audio quality might not be as pristine as with the original WAV. But for most everyday listening, this trade-off between file size and audio quality is more than acceptable.

### Streaming Audio: A River of Sound

Now, let's consider a situation that's probably familiar to most of you: streaming audio. When you play a song on Spotify, Apple Music, or another streaming service, the audio doesn't arrive all at once. Instead, it's sent over the internet in a steady stream of data that your device plays as it arrives. This is, in fact, another form of compression!

In contrast to formats like MP3 that must be fully downloaded before you can listen, streaming formats are designed to start playing after downloading only a small portion of the file. They're typically compressed even further than MP3s to minimize data usage and ensure uninterrupted playback, even on slower internet connections.

Behind the scenes, complex algorithms dynamically adjust the quality of streaming audio based on your network conditions. Ever noticed how the audio quality dips when your Wi-Fi signal is weak but improves once the connection gets better? That's these algorithms at work!

So, the next time you're grooving to your favorite tunes, spare a thought for the magic of audio files. Whether it's the unaltered symphony of a WAV, the clever compression of an MP3, or the dynamic stream of online music, remember: it's all just the transformation of bytes into beautiful sound!

## Squishing Data: The Magic of Compression

Ever tried to pack for a long trip with a suitcase that seems too small? You start by neatly folding your clothes, but soon you're sitting on the suitcase, trying to zip it up while shoving socks into every nook and cranny. That's a bit like what we do with data when we use compression!

Compression is a technique to make files smaller by reducing the amount of data they contain. It's a fundamental concept in computer science and is used in many areas, from making websites load faster, to streaming music, to sending space probes to other planets!

- **Lossless compression**, as the name implies, is a way to reduce the size of a file without losing any information. It's like folding your clothes more efficiently to make them fit into your suitcase. When you unpack (or decompress) the file, you'll find everything exactly as it was. No detail is lost!

  One of the most common forms of lossless compression is the **ZIP** file format. Have you ever downloaded a bunch of files bundled together in a .zip file? That's lossless compression at work! Programs, documents, images - anything can be zipped. It's like a suitcase for your files!

- **Lossy Compression.** Now, imagine you're packing for your trip and you decide you don't really need seven pairs of shoes. You can make do with just two. By leaving some shoes at home, you've made a lot more space in your suitcase. That's similar to lossy compression.

  Lossy compression reduces file size by discarding less important data. When you decompress the file, it won't be exactly the same as the original, but if done well, the changes won't be noticeable. We use lossy compression when a perfect copy isn't necessary, like in MP3 music files or JPEG images.

Just like with packing your suitcase, there are trade-offs with compression. Lossless compression won't reduce the file size as much as lossy compression, but you won't lose any data. Lossy compression can make files a lot smaller, but you can't get back the data that's been discarded. It's like leaving your favourite shoes at home -- once you're on your trip, you can't suddenly decide you want to wear them after all!

Understanding when to use each type of compression is a vital part of working with data. Sometimes, you need every detail, so lossless compression is the way to go. Other times, you just need the general idea (like when you're streaming a video), so lossy compression is a better choice.

## Summary Table: Common File Formats
Here's a table summarizing major file formats:

| Category | File Format/Extension | Summary |
| --- | --- | --- |
| Source Code Files | `.py`, `.java`, `.c`, `.cpp`, `.js`, `.html` | These are text-based files containing programming code. They can be read and edited with any text editor but are designed to be interpreted or compiled by a programming language processor. |
| Data Files | `.csv`, `.json`, `.xml`, `.sql`, `.sqlite` | These formats are used for storing structured data. They are often used for data transfer between different software applications or systems. |
| Audio Files | `.wav`, `.mp3`, `.aac`, `.flac`, `.ogg` | These formats are used to store digital audio data. The encoding can be lossy (like `.mp3`, `.aac`) or lossless (like `.wav`, `.flac`). |
| Image Files | `.jpg`, `.png`, `.bmp`, `.gif`, `.tiff` | These formats are used to store digital images. They can use either lossy or lossless compression, and some support transparency. |
| Video Files | `.mp4`, `.avi`, `.mov`, `.flv`, `.wmv` | These are formats for storing digital video data. They differ in compression algorithms and supported features (like subtitles or multiple audio streams). |
| Document Files | `.docx`, `.xlsx`, `.pptx`, `.pdf`, `.txt`, `.md` | These formats are used to store different types of documents: text documents, spreadsheets, presentations, PDFs, etc. They can contain text, images, tables, and other elements. |
| Executable Files | `.exe`, `.app`, `.dmg`, `.sh`, `.bat` | These files contain programs in a format that a computer's operating system can directly execute. |
| Compressed Files | `.zip`, `.rar`, `.tar.gz`, `.7z` | These formats are used to store one or more other files in a compressed form to save disk space and to make downloading or transferring easier. |
| Database Files | `.db`, `.mdb`, `.accdb`, `.dbf` | These formats are used by various database management systems to store structured data. They are not usually meant to be opened or edited directly by the user. |
| Markup Language Files | `.html`, `.xml`, `.json`, `.yaml` | These are text-based formats used to store structured data. They're often used in web development and data interchange. |

Remember that some files can fit in more than one category. For instance, `.xml` and `.json` can be used to store data (thus fitting into the "Data Files" category) but are also markup languages (thus fitting into the "Markup Language Files" category).

## Glossary
| Term | Definition |
| --- | --- |
| Computer file | A unit of digital data created by a software application, stored on a storage medium like a hard drive, and identifiable by a unique name or path. |
| File format | A design for the layout of a file's data, specifying how the bits encode information and determine how a file is interpreted by software. |
| File extension | The part of a filename, usually appearing after a period, that indicates the type of data stored in the file. |
| ASCII | A standard for character encoding which encodes 128 specified characters into seven-bit integers. |
| Unicode | A character encoding standard designed to support the digital processing and display of the written texts of all languages. |
| Markup Language | A system for annotating text that surrounds the text with codes to indicate structure, semantic meaning, or presentation formatting. |

| XML | Extensible Markup Language, a text-based data format with strong support via Unicode for different human languages. |

| Compiled language | A programming language that is typically implemented using compilers, i.e., translators that generate machine code from source code. |
| Executable file | A file that can perform specific operations or tasks as it contains a program capable of being run independently. |
| Assembly language | A programming language corresponding closely to machine code but allowing symbolic representation of numeric values and memory locations. |
| Machine code | The lowest level of programming language, executed directly by the CPU, composed of binary instructions. |

| MP3 | A popular audio coding format for digital audio which uses a form of lossy data compression to reduce file sizes. |

| Lossless compression | A type of compression where no data is lost when the compressed file is decompressed, ensuring it's exactly the same as the original. |
