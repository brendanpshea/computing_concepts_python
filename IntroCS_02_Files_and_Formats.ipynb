# Plain Text Files (txt), ASCII, and Unicode
**Plain text files** are among the simplest kinds of computer files. They contain only text characters, without any special formatting or metadata (like bold, italics, or hyperlinks). The most common plain text file format has the extension '.txt'. 

Plain text files are incredibly versatile. You can use them to jot down notes, write code, or even store numerical data. But how does your computer transform a sequence of 1s and 0s into meaningful text? The magic happens through character encoding standards like ASCII and Unicode.

### ASCII: The Basic Building Block

ASCII (American Standard Code for Information Interchange) was one of the first widely accepted standards for translating numbers into text. It uses seven bits to represent each character. Because each bit can be either a 0 or a 1, seven bits can represent 2^7 (128) different values.   Here's a tiny slice of the ASCII table to give you an idea:

| Character | Binary | Decimal | Hexadecimal |
| --- | --- | --- | --- |
| 'A' | 01000001 | 65 | 41 |
| 'a' | 01100001 | 97 | 61 |
| '3' | 01100011 | 51 | 33 |
| '+' | 00101011 | 43 | 2B |
| '@' | 01000000 | 64 | 40 |
| '$' | 00100100 | 36 | 24 |
| '%' | 00100101 | 37 | 25 |
| '?' | 00111111 | 63 | 3F |
| '(' | 00101000 | 40 | 28 |
| ')' | 00101001 | 41 | 29 |

Binary is the numeric representation of the character in binary, decimal is the same value in decimal, and character is the equivalent character.

ASCII can encode 128 different characters in total: 33 control characters (such as carriage return, line feed, and tab), and 95 printable characters including digits (0-9), uppercase and lowercase letters (A-Z, a-z), and special symbols (like !, @, #, $, %, &, *, (, ), and ?).

### Executable Files: More Than Meets the Eye

An executable file is a special kind of file that is capable of performing tasks according to encoded instructions. Instead of just storing information, like a text document or an image, an executable file is a program that's ready to be launched and perform its task on a computer.

Executable files come with various extensions, depending on the operating system and the type of program they're associated with. For instance:

-   On Windows, executable files usually have an `.exe` extension. When you play a game on your Windows computer, the game is typically an `.exe` file.

-   On macOS, executable files usually have a `.app` extension. So when you open the Safari web browser, you're actually running the `Safari.app` executable file.

-   On Linux, executable files don't necessarily have a specific extension, but a common one is `.bin`.

### The Magic Behind the Click

The executable files that we interact with daily are often the result of compiled source code. Developers write source code in high-level programming languages like C++, Java, or Python. This source code is then fed into a compiler or an interpreter that translates it into machine code. This machine code is then bundled into an executable file, waiting for you to click and unleash its magic.

In essence, whenever you open a program, play a game, or use a software application, you're running an executable file. The process of creating these files from source code is a fundamental concept in computer science, much like following a recipe is fundamental in cooking. ## Compilation: The Transformation Process

When you write source code in a language like C (the "mother language" of modern languages like Python and C, which is still commonly used to write low-level software programs), it's essentially a set of high-level instructions that humans can understand. However, your computer's processor doesn't speak C---it understands only binary machine code. This is where a compiler comes in.

The **compiler** is a program that reads your source code and translates it into machine code, a low-level language that your computer's processor can understand. This machine code is then packaged into an executable file.  Let's consider a simple C program:

```
// A program written in C

#include <stdio.h> // Library for standard input and output

int main() {
   // printf() displays the string inside the quotation
   printf("Hello, World!");
   return 0; // This means "exit without error"
}
```

When you compile this program, the C compiler (like gcc) processes it into machine code.

### Assembly and Machine Code: A Glimpse into the Binary World

Under the hood, the compilation process typically involves an intermediate step where the high-level instructions are first translated into **assembly language**---a more human-readable form of machine language.

For example, the `printf` function in our C program might be translated into an assembly instruction like this:

```
; A sample program written in "assembly language"


mov edi, offset .L.str   ; address of "Hello, World!" string
call puts                ; call puts function to print the string
```

Each of these assembly instructions is then translated into binary **machine code**, like `10110000 01110011 01100001`, which is what the processor actually understands and executes.

Though it might seem overwhelming at first, the key takeaway is that different numbers (binary digits in this case) can correspond to different instructions based on their context---a fundamental concept in computer science.

### Running Python: An Interpreted Language

Python, unlike C, is an interpreted language. That means instead of being compiled once into an executable file, Python code is translated on-the-fly each time it is run. The program that does this translation is the Python interpreter, which is itself an executable file that was compiled from C source code!

When you run a Python program, the Python interpreter reads your Python code, line by line, and translates each line into machine code instructions that your computer's processor can understand.  So, even though you're writing Python code, the magic of the Python interpreter (a product of compiled C code) is working behind the scenes to execute your Python programs. It's like having an instant translator facilitating a conversation between you (speaking Python) and your computer (speaking machine language).

Think about the last time you visited a library. Rows of books neatly arranged, each carrying unique information, systematically classified and labeled for easy access. Data files, in the world of computer science, are akin to these books. They are specifically structured to store, organize and retrieve information efficiently. Among the numerous types of data files, we'll focus on two major types you'll encounter later in the class: Comma-Separated Values (CSV) files and SQLite database files.

### CSV Files: The Notepad of Data Storage

Consider the simplest form of storing data - a CSV file. The acronym 'CSV' stands for 'Comma-Separated Values', which precisely captures the essence of this file type: it contains values separated by commas.

```
Student_Name,Major,Credits_Completed
```

In this CSV file, each line corresponds to a different college student, and the comma-separated values represent different attributes for each student: their name, major, and the number of completed credits. CSV files are nearly universally accepted across platforms and programming languages due to their simplicity, making them an excellent choice for storing and transferring straightforward tabular data.

### SQLite Database Files: The Encyclopedia of Data Management

As we venture into more complex and large-scale data, we require a more structured and powerful system: a database. SQLite, an embedded database engine, employs a self-contained, serverless design and requires no additional setup, making it an excellent choice for many applications. It stores data in a file with the extensions `.sqlite` or `.db`.  SQLite is one of many database file formats, with others including MySQL (.frm), PostgreSQL (.mkd), and Oracle Database (.dbf), each having its own strengths and use cases.


A single SQLite database file, unlike a CSV, can contain multiple tables, indexes, triggers, and views, allowing it to efficiently manage complex relational data.  Let's imagine a college student data scenario:

```
Database: University.db

Table: Students
-------------------------------------
| ID | Name    | Major | Credits |
-------------------------------------
| 1  | John    | CS    | 60      |

-------------------------------------

Table: Courses
-------------------------------------
| ID | Name          | Professor  |
-------------------------------------
| 1  | Data Structures | Dr. White |

-------------------------------------
```

Here, the SQLite database 'University.db' consists of two tables, 'Students' and 'Courses', each storing information in rows (entries) and columns (attributes).

### Binary Format: The Secret Behind Database Files

Unlike CSV files or other text-based data files, database files like SQLite are stored in a binary format. This means that if you tried to open an SQLite file in a text editor, you wouldn't see neatly arranged rows and columns of data. Instead, you would see a seemingly incomprehensible mix of characters, some even unprintable. That's because the binary data is not meant to be human-readable; it's structured to be efficiently read and manipulated by a machine.

- This binary data is organized in a very specific way. Each SQLite database file, for example, is divided into **"pages."** Each page has a fixed size (typically 4096 bytes), and it's this rigid structure that allows SQLite to quickly locate and access data, no matter how large the database becomes.

- The way the data is stored in pages also provides for more efficient use of storage space. In a text file like CSV, each character consumes one byte of space, regardless of its significance. But in a SQLite file, the more compact binary format allows for more data to be stored in less space.

- Each type of database file format will have its own unique method of arranging binary data. So, SQLite database files can't be opened by MySQL or Postgres. However, they all have a similar "logic."

It's this complex organization of binary data that allows database files to handle more complex operations, be more efficient, and provide more robust data management capabilities compared to text-based formats.

### Beyond Text: Advantages of Database Files

SQLite, like many database file formats, offers several advantages beyond simple text-based formats:

1.  Efficiency & Speed: Databases are optimized for large data sets and provide fast access, search, and update of data.
2.  Reduced Storage: Database systems can compress data, reducing the amount of physical storage space required.
3.  Security: Databases often come with built-in security measures, such as encryption and access control.
4.  Data Integrity: Databases maintain a consistent state of data even in case of system failures.

However, these advantages must be weighed against the strengths of CSVs (they are highly portable between different systems, and require no special training/software to set up). However, for many software and business applications, database files are the "rule."

## Picture This: Image Data in a Digital World

To a computer, an image is simply an array of numbers, each representing a color. But to us, it forms a picturesque scene, a memory, or a piece of art. How does this transformation happen? Let's dive into the world of image data.

### RGB Values: The Primary Colors of the Digital World

Images are comprised of tiny individual units called pixels, each pixel corresponding to a specific color. This color is often represented as an **RGB value**, an acronym for Red, Green, Blue.  RGB values are tuples of three numbers, each ranging from 0 to 255. For example, the RGB value of (255, 0, 0) represents pure red, (0, 255, 0) is pure green, and (0, 0, 255) is pure blue. By combining these primary colors in varying amounts, we can represent millions of different colors. The RGB value (0, 0, 0) represents black, while (255, 255, 255) represents white.

### Bitmap (BMP) Files: A Pixel-by-Pixel Story

The **Bitmap (BMP)** format is one of the simplest image file formats. It was developed by Microsoft for use on personal computers. The BMP format stores color information for each pixel in the image without any compression. This means that the file size of a BMP image can be quite large, but it also means that the image quality is high, as no information is lost.

For a simplistic understanding, imagine a 3x3 pixel image with RGB values as follows:

````python
import matplotlib.pyplot as plt
import numpy as np

# A 3x3 array for colored image
# each set of three numbers represents red, green, and blue
color_image = np.array([[[255, 0, 0],   [0, 255, 0],   [0, 0, 255]],
                        [[255, 255, 0], [0, 255, 255], [255, 0, 255]],
                        [[0, 0, 0],     [255, 255, 255],[127, 127, 127]]], dtype=np.uint8)

# Display the image
plt.imshow(color_image)
plt.axis('off') # to hide the axes
plt.show()
````

### JPEG Files: Picture Perfect Compression

**JPEG**, standing for Joint Photographic Experts Group (the team that developed the format), is a commonly used method of lossy compression for digital images. If you've ever taken a photo with a digital camera or a smartphone, you've likely used the JPEG format.

Unlike BMPs, JPEGs undergo a process of compression where some data is discarded to reduce the file size. This process involves a complex transformation of the image data into a format that can be stored more compactly, at the expense of some loss of image quality.

For example, imagine that same 3x3 pixel image we used before. In the BMP version, we store the exact RGB value of each pixel, which can take a lot of space for large images. In the JPEG version, the image might be transformed and compressed so that the color information is stored in a way that requires less space.

The amount of compression in a JPEG file can be adjusted. Higher levels of compression result in lower quality images but smaller file sizes, while lower levels of compression result in higher quality images but larger file sizes.

The JPEG format is incredibly popular for digital photography and online images due to its balance of quality and file size. Understanding how it works is a perfect example of how clever algorithms and understanding of human perception can lead to powerful tools in computer science.

````python
# Example showing the difference between BMP and JPEG
# Requires running previous cell
from PIL import Image
import os
import numpy as np

np.random.seed(42)  # for reproducibility

# Create a 100x100 array with random shades of blue
shades_of_blue = np.zeros((100, 100, 3), dtype=np.uint8)
shades_of_blue[:, :, 2] = np.random.randint(0, 256, size=(100, 100))  # Random blue values

# Create a PIL Image object from the NumPy array
img = Image.fromarray(shades_of_blue)

# Save the image as BMP and JPEG
img.save('shades_of_blue.bmp')
img.save('shades_of_blue.jpg', 'JPEG')

# Display file information
print_file_info('shades_of_blue.bmp')
print_file_info('shades_of_blue.jpg')
````

## Listening to the Bytes: Audio Files

Music, podcasts, voice chats, or your favorite YouTube videos - audio is a fundamental part of our digital experience. But how do our devices transform sequences of numbers into something that makes sense to our ears? Let's dive into the symphony of audio file formats, with a special focus on WAV, MP3, and the magic behind streaming audio.

### WAV Files: The Unaltered Symphony

Imagine you're attending a live concert. The musicians strike their instruments and the sound waves begin their journey towards your ears. If we could capture these sounds in "freeze frames" of time, each snapshot would show a complex pattern of waves. This process forms the essence of digital audio.

In a WAV file (Waveform Audio File Format), these audio 'snapshots' are stored in their raw, uncompressed form. Each of these snapshots, known as 'samples', records the strength (amplitude) of the sound wave at a specific moment in time. When these samples are played back rapidly - around 44,100 times a second for CD quality sound - your brain perceives it as continuous sound, almost indistinguishable from the live performance!

However, this high fidelity comes with a cost: file size. Since WAV files store all these samples without any compression, they can take up quite a bit of space. For instance, just a minute of CD-quality stereo sound can require more than 10 MB!

### MP3 Files: Clever Compression

What if you want to share your concert recording with friends over the internet? Uploading or downloading a full 60-minute WAV file would mean transferring over 600 MB, which could be time-consuming or even impossible with slower connections. That's when the MP3 format, a superstar of compressed audio, comes to our rescue.

MP3 files apply a smart trick - they eliminate the parts of the audio that our ears typically can't hear. This process, known as 'lossy compression', substantially shrinks the file size while keeping the perceived audio quality high. The same 60-minute recording that was 600 MB in WAV format might only be about 60 MB as an MP3!

However, it's important to note that some audio information is lost during this compression, so the audio quality might not be as pristine as with the original WAV. But for most everyday listening, this trade-off between file size and audio quality is more than acceptable.

### Streaming Audio: A River of Sound

Now, let's consider a situation that's probably familiar to most of you: streaming audio. When you play a song on Spotify, Apple Music, or another streaming service, the audio doesn't arrive all at once. Instead, it's sent over the internet in a steady stream of data that your device plays as it arrives. This is, in fact, another form of compression!

In contrast to formats like MP3 that must be fully downloaded before you can listen, streaming formats are designed to start playing after downloading only a small portion of the file. They're typically compressed even further than MP3s to minimize data usage and ensure uninterrupted playback, even on slower internet connections.

Behind the scenes, complex algorithms dynamically adjust the quality of streaming audio based on your network conditions. Ever noticed how the audio quality dips when your Wi-Fi signal is weak but improves once the connection gets better? That's these algorithms at work!

So, the next time you're grooving to your favorite tunes, spare a thought for the magic of audio files. Whether it's the unaltered symphony of a WAV, the clever compression of an MP3, or the dynamic stream of online music, remember: it's all just the transformation of bytes into beautiful sound!

## Squishing Data: The Magic of Compression

Ever tried to pack for a long trip with a suitcase that seems too small? You start by neatly folding your clothes, but soon you're sitting on the suitcase, trying to zip it up while shoving socks into every nook and cranny. That's a bit like what we do with data when we use compression!

Compression is a technique to make files smaller by reducing the amount of data they contain. It's a fundamental concept in computer science and is used in many areas, from making websites load faster, to streaming music, to sending space probes to other planets!

## Zip It Up: Lossless Compression

Lossless compression, as the name implies, is a way to reduce the size of a file without losing any information. It's like folding your clothes more efficiently to make them fit into your suitcase. When you unpack (or decompress) the file, you'll find everything exactly as it was. No detail is lost!

One of the most common forms of lossless compression is the ZIP file format. Have you ever downloaded a bunch of files bundled together in a .zip file? That's lossless compression at work! Programs, documents, images - anything can be zipped. It's like a suitcase for your files!

## Let It Go: Lossy Compression

Now, imagine you're packing for your trip and you decide you don't really need seven pairs of shoes. You can make do with just two. By leaving some shoes at home, you've made a lot more space in your suitcase. That's similar to lossy compression.

Lossy compression reduces file size by discarding less important data. When you decompress the file, it won't be exactly the same as the original, but if done well, the changes won't be noticeable. We use lossy compression when a perfect copy isn't necessary, like in MP3 music files or JPEG images.

## Trade-Offs

Just like with packing your suitcase, there are trade-offs with compression. Lossless compression won't reduce the file size as much as lossy compression, but you won't lose any data. Lossy compression can make files a lot smaller, but you can't get back the data that's been discarded. It's like leaving your favourite shoes at home -- once you're on your trip, you can't suddenly decide you want to wear them after all!

Understanding when to use each type of compression is a vital part of working with data. Sometimes, you need every detail, so lossless compression is the way to go. Other times, you just need the general idea (like when you're streaming a video), so lossy compression is a better choice.
